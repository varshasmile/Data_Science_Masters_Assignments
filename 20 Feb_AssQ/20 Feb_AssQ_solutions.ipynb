Q1. Explain GET and POST methods.

GET Method:

GET is a request method used to retrieve data from a specified resource.
When a client sends a GET request, it appends the data as part of the URL query parameters.
GET requests are typically used to retrieve data, such as fetching a webpage or accessing a specific resource.
They are considered safe and idempotent, meaning they should not have any side effects on the server and can be repeated without changing the result.
GET requests can be cached by browsers and are visible in the URL, making them less secure for sensitive data.

POST Method:

POST is a request method used to submit data to be processed by a specified resource.
When a client sends a POST request, it includes the data in the body of the request, which is not visible in the URL.
POST requests are typically used to send data to a server for processing or to create/update a resource.
They can have side effects on the server, such as modifying data or triggering actions.
POST requests are not cached by browsers and are more secure for transmitting sensitive data compared to GET requests.

Q2. Why is request used in Flask?

The request object in Flask is used to access and handle incoming HTTP requests made by clients (such as web browsers) to the Flask application. It provides access to various properties and methods that allow developers to interact with the request data.

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client's browser to a different URL. It is commonly used in web applications to redirect users to a different page or route after performing a certain action or when a specific condition is met.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to separate the presentation (HTML) from the application logic (Python code). Templates allow developers to create dynamic web pages by embedding placeholders and expressions that are replaced with actual data when the page is rendered.

The render_template() function in Flask is used to render a template and generate the HTML content that will be sent back to the client's browser. It takes the name of the template file as an argument and can also accept additional variables that will be passed to the template for rendering.

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/postman_api/hello', methods=['GET'])
def hello():
    return jsonify(message='Hello, World!')

@app.route('/postman_api/greet', methods=['POST'])
def greet():
    data = request.get_json()
    name = data.get('name')
    return jsonify(message=f'Hello, {name}!')

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
