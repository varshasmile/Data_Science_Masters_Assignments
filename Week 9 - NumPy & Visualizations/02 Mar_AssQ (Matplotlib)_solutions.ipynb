Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of functionalities for creating static, animated, and interactive visualizations. It is extensively used for generating high-quality plots, charts, histograms, and other types of visual representations of data.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a basic plot that represents data points connected by straight lines. It is commonly used to visualize the trend or relationship between two variables over a continuous interval.

Scatter Plot: A scatter plot displays individual data points as markers on a two-dimensional plane. It is useful for showing the distribution and correlation between two variables.

Bar Plot: A bar plot, also known as a bar chart, uses rectangular bars to represent categorical data. It is often used to compare different categories or groups.

Histogram: A histogram is used to visualize the distribution of a single variable. It divides the data into bins and displays the frequency or count of data points falling into each bin.

Pie Chart: A pie chart represents data as slices of a circular pie. It is useful for showing the composition or proportion of different categories in a dataset.

A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional plane. It is useful for visualizing the distribution and relationship between two continuous variables.

```python
import matplotlib.pyplot as plt
import numpy as np 

np.random.seed(3) 

x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x))
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to create multiple plots in a single figure window, arranged in a grid-like structure. Each subplot can display a different plot type or represent a different subset of data.

```python
import matplotlib.pyplot as plt
import numpy as np 

# For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500]) 
```

```python
# For line 2: 
y2 = np.array([50, 20, 40, 20, 60, 70]) 
```

```python
# For line 3: 
y3 = np.array([10, 20, 30, 40, 50, 60]) 
```

```python
# For line 4: 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Creating subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Plot 1
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Line 1')

# Plot 2
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Line 2')

# Plot 3
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Line 3')

# Plot 4
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Line 4')

# Adjusting spacing
plt.tight_layout()

# Displaying the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot that uses rectangular bars to represent categorical data. It is commonly used to compare and display the values of different categories or groups.

Bar plots are useful for:

Comparing Values: Bar plots allow easy visual comparison of values between different categories.

Showing Distribution: Bar plots can also be used to represent the distribution of data within each category, by displaying stacked or grouped bars.

Presenting Rankings: Bar plots are effective in displaying rankings or sorting data based on specific criteria.

```python
import matplotlib.pyplot as plt
import numpy as np 

company = np.array(["Apple", "Microsoft", "Google", "AMD"]) 
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Bar plot
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()
```

```python
# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200) 
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays the five-number summary of a dataset, including the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values. It is used to identify the central tendency, spread, skewness, and potential outliers within a dataset.

Box plots are useful for:

Comparing Distributions: Box plots allow for easy comparison of distributions between different groups or categories. By placing multiple box plots side by side, you can visually compare the central tendency and spread of different datasets.

Detecting Outliers: Box plots display potential outliers as individual data points or "fliers" outside the whiskers. Outliers can provide insights into unusual or extreme observations that deviate significantly from the bulk of the data.

Identifying Skewness: Box plots can provide an indication of the skewness of a distribution. Skewed distributions tend to have asymmetric box plots, where the whisker lengths and positions differ on either side of the box.

```python
data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
